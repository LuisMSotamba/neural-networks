```python
from PIL import Image
from tqdm import tqdm
import os

def find_corrupt_images(root_dir):
    print(f"Scanning {root_dir} for corrupt images...")
    bad_files = []
    
    # Walk through all directories
    for root, _, files in os.walk(root_dir):
        for file in tqdm(files, desc="Checking files"):
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                file_path = os.path.join(root, file)
                try:
                    # Try to open and fully load the image
                    with Image.open(file_path) as img:
                        img.load() 
                except OSError:
                    print(f"\nFOUND BAD FILE: {file_path}")
                    bad_files.append(file_path)
                except Exception as e:
                    print(f"\nOther error with {file_path}: {e}")

    return bad_files

# Replace with the path to your actual images
bad_images = find_corrupt_images("/home/luchocode/projects/tesis/data/selected_exoimages/train")
print(f"Total corrupt images found: {len(bad_images)}")
```
